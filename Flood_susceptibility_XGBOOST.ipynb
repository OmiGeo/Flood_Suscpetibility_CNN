```python
x = df[['Soil_','Slope', 'Rainfall1', 'NDVI', 'lulc_', 'Geomorphol', 'Geology_',
       'Flow_acc', 'DR_final', 'dem', 'DD_final', 'Curvature', 'CN', 'Aspect',
       'TWI', 'STI', 'SPI']]
y = df['F']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(df.corr(), annot =True)
plt.title('Correlation Matrix')
```

```python

# import modules
import matplotlib.pyplot as mp
import pandas as pd
import seaborn as sb
```

```python
dataplot = sb.heatmap(x.corr(), cmap="YlGnBu", annot=True)
mp.figure(figsize=(10,20))
# displaying heatmap
mp.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
object= StandardScaler()
x_scale = object.fit_transform(x) 
print(x_scale)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_scale,y,test_size=0.3, random_state= 42)
```

```python
! pip install xgboost
```

```python
import numpy as np
```

```python
import xgboost as xgb
```

```python
xgb_classifier = xgb.XGBClassifier(eta = 0.05)
model = xgb_classifier.fit(x_train, y_train)
    
```

```python
prediction = model.predict(x_test)
prediction
```

```python
prediction_prob=model.predict_proba(x_test)
prediction_prob
```

```python
from sklearn.metrics import plot_roc_curve
```

```python
import matplotlib.pyplot as plt
```

```python
ax = plt.gca()
model_disp = plot_roc_curve(model, x_test, y_test, ax=ax, alpha=0.8, color='red')
plt.xlabel('1-Specificity', fontsize=16)
plt.ylabel('Sensitivity', fontsize=16)
plt.xticks(fontsize= 12)
plt.yticks(fontsize= 12)
plt.title('ROC curve', fontsize=16)
plt.show()
```

```python
from sklearn import metrics
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, prediction)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

```python
 from sklearn.metrics import classification_report
```

```python
features_list = list(x.columns)
```

```python
feature_imp = pd.Series(model.feature_importances_, index= features_list ).sort_values(ascending=False)
print(feature_imp)
```

```python
Accuracy = metrics.accuracy_score(y_test, prediction)
Accuracy
```

```python
Precision = metrics.precision_score(y_test, prediction)
Precision
```

```python
Sensitivity_recall = metrics.recall_score(y_test, prediction)
Sensitivity_recall
```

```python
Specificity = metrics.recall_score(y_test, prediction)
Specificity
```

```python
F1_score = metrics.f1_score(y_test, prediction)
F1_score
```

```python
print({"Accuracy":Accuracy,"Precision":Precision,"Sensitivity_recall":Sensitivity_recall,"Specificity":Specificity,"F1_score":F1_score})
```
